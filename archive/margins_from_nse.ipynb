```python
#***       NSE margin data for Equities       ***
# ______________________________________________

# Margins for equities extracted from the web-site:
# https://www.nseindia.com/archives/nsccl/var/C_VAR1_ddmmyyyy_1.DAT

import datetime
import requests
import pandas as pd

# Set date in the right format for today
dt = datetime.datetime.now().strftime('%d%m%Y')

url = 'https://www.nseindia.com/archives/nsccl/var/C_VAR1_'+dt+'_1.DAT'
page = requests.get(url).text

data = page.replace('\n', ';') # replaces the newline feeds

# Remove the first row, and split the data into columns
df_raw = pd.read_csv(pd.compat.StringIO(data), header=None, sep=';').T[1:][0].str.split(',', expand=True)
df_raw = df_raw.rename({0: 'RecType', 1: 'Symbol', 2: 'Series', 
               3: 'ISIN', 4: 'SecVaR', 5: 'IndexVaR',
               6: 'VaRMargin', 7: 'XLossRate', 8: 'AdhocMargin', 9: 'NSEMarginRate'}, axis=1)

# Convert the data into float %ge
fields = 'SecVaR,IndexVaR,VaRMargin,XLossRate,AdhocMargin,NSEMarginRate'.split(',')
df_raw[fields] = df_raw[fields].apply(pd.to_numeric).div(100)

df = df_raw[['Symbol', 'NSEMarginRate']]
df
```
