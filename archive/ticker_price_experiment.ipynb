```python
# Sample program for getting tick prices
from ib_insync import *
util.startLoop()
ib = IB().connect('127.0.0.1', 4001, clientId=15) # IBG live
```

```python
import datetime
symbol = 'ACC'
contract = Stock(symbol, exchange='NSE')
ticker = ib.reqMktData(contract, snapshot=True)
(ticker, datetime.datetime.now())
```

```python
while ticker.close != ticker.close:
    ib.sleep(0.01)
    print(ticker, datetime.datetime.now())
```
